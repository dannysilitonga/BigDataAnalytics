# LINEAR REGRESSION 

Interest Rate = b1*(FiCO Score) + b2*(Loan Amount)

```python
intrate = loansData['Interest.Rate']
loanamt = loansData['Amount.Requested']
fico    = loansData['FICO.Score']
```

When a column from a dataFrame is extracted, it is returned as a Series datatype.  So we need to reshape the data as follows:

```python
# The dependent variable
y = np.matrix(intrate).transpose()
# The independent variables shaped as columns
x1 = np.matrix(fico).transpose()
x2 = np.matrix(loanamt).transpose()
```

```python
# Put together the 2 columns, creating an input matrix
x = np.column_stack([x1, x2])
```

```python
# Create a linear model
X     = sm.add_constant(x) 
model = sm.OLS(y,X)
f     = model.fit()
```

```python
print 'Coefficients:', f.params[1:3]
print 'Intercept:', f.params[0]
print 'P-Values:', f.pvalues 
print 'R-Squared:', f.rsquared
```

The closer the p-value to 0, the more confident one should be that we have a good estimate. According to R^2, this model has explained 65% of the variance in the data
